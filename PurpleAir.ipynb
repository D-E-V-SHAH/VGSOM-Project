# Data Handling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
data_high = pd.read_csv('MultiSensor_IntercalibrationData.csv')
data_low = pd.read_csv('SingleSensor_CalibData.csv')
data_low
```

```python
data_high
```

```python
data_high_required = data_high.loc[:, ['Time', 'S_2_PM2.5A']]
data_high_required
```

```python
data_low_required = data_low.loc[:, ['Time_Beirut', 'PM2.5 ConHR(ug/m3)']]
data_low_required
```

```python
data_high_required['Time'] = pd.to_datetime(data_high_required['Time'])
data_high_required
```

```python
data_low_required['Time_Beirut'] = pd.to_datetime(data_low_required['Time_Beirut'])
data_low_required
```

```python
index = []
for indexi, row in data_high_required.iterrows():
    if (row['Time'] in data_low_required.Time_Beirut.values) == True:
        index.append(indexi)
        
data_1 = data_high_required.loc[index, :]
```

```python
data_1.reset_index(inplace = True)
data_1.drop(['index'], axis=1)
```

```python
index = []
for indexi, row in data_low_required.iterrows():
    if (row['Time_Beirut'] in data_high_required.Time.values) == True:
        index.append(indexi)
        
data_2 = data_low_required.loc[index, :]
data_2
```

```python
data = pd.DataFrame(data_1.Time)
data['Predicted'] = data_2['PM2.5 ConHR(ug/m3)']
data['Actual'] = data_1['S_2_PM2.5A']
data
```

```python
data['Predicted'].plot(label='Predicted', color='orange')
data['Actual'].plot(label='Actual', color='green')
plt.show()
```

# Model

```python
window_size = 6
```

```python
x = data['Predicted'].iloc[:window_size]
x = x.to_numpy()
X = []
y = []
for i in range((len(data) - window_size - 1)):
    row = [[a] for a in x[i:i+window_size]]
    X.append(row)
    x = np.concatenate((x, [data['Actual'][i + window_size + 1]]))
    y.append(x[-1])
y = np.array(y)
X = np.array(X)
print(X.shape, y.shape)
```

```python
Train_X = X[:120]
Train_y = y[:120]
Val_X = X[120:150]
Val_y = y[120:150]
Test_X = X[150:]
Test_y = y[150:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((window_size, 1)))
model1.add(LSTM(32))
model1.add(Dense(16, activation='relu'))
model1.add(Dense(1, activation='linear'))

model1.summary()
```

```python
model1.fit(Train_X, Train_y, validation_data = (Val_X, Val_y), epochs = 200)
```

```python
train_predictions = model1.predict(X).flatten()
train_results = pd.DataFrame(data={'Train Predictions':train_predictions, 'Actuals':y})
```

# Evaluating Error

```python
import math

avg_error = 0
mse_error = 0
for i in range(len(train_results) - len(Test_X), len(train_results) - 2):
    avg_error += math.sqrt((train_results['Train Predictions'][i+1] - train_results['Actuals'][i]) ** 2)
    mse_error += (train_results['Train Predictions'][i+1] - train_results['Actuals'][i]) ** 2
#     print(avg_error)
avg_error /= len(Test_X)
mse_error /= len(Test_X)
print(f"The average error on PurpleAir dataset is {avg_error}")
print(f"The mean sqaured error on PurpleAir dataset is {mse_error}")
```
